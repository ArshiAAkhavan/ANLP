<div dir="rtl">
    <h2>
        بخش سوم - استخراج واحدهای اندازه‌گیری
    </h2>
    
    این بخش از دو بخش اصلی <code> tagging </code> و <code> pattern matching </code> تشکلیل شده است.
    <h3> tagging </h3>
    در این بخش به‌کمک ابزار‌های مختلفی مانند <code> parsi_io </code> و <code> hazm </code> و همچنین به کمک داده‌هایی که از <code> unit_retriever </code> بدست می‌آید، اجزای مختلف جمله از قبیل <code> number </code> , <code> unit </code> , <code> quantity </code> , <code> stop_word </code> , <code> adverb </code> را نشانه گذاری می‌کنیم.
    
    <h3> pattern matching </h3>
    حال که متن ورودی را نشانه گذاری کرده‌ایم و اجزای مختلف آن را شناسایی نمودیم آن را با لیستی از <code> pattern </code> هایی که داریم مقایسه می‌کنیم و واحد‌های مورد نظر را از متن استخراج می‌کنیم
    
    تعدادی از <code> pattern </code> های مجاز در برنامه:
    <code>
    <c
    
    <br>
    ابتدا مثال‌های تمرین را حل می‌کنیم:
</div>

````python
from pprint import pprint
from dataclasses import asdict
from unit_extractor import UnitExtractor

extractor = UnitExtractor()
````

````python
texts = [
    'علی ۳.۵ کیلوگرم آرد خرید و باتری خود را هشتاد و پنج صدم وات شارژ کرد.',
    'شهاب‌سنگی به تندی ۱۵ km/s وارد جو زمین شد.',
    'یک خودرو با سرعت زیاد از ما سبقت گرفت',
]

for text in texts:
    print(text)
    pprint([asdict(result) for result in extractor.run(text)], sort_dicts=False)
    print('\n'.join(['*' * 50] * 3))
    print()
````

<div dir="rtl">
حال برای داده‌ی دریافت شده از ویکی‌پدیا، واحدهای اندازه‌گیری را استخراج می‌کنیم.
</div>

````python
from tqdm import tqdm

results = []
for document in tqdm(data):
    r = extractor.run(document['text'])
    r = [asdict(result) for result in r]
    results.append(dict(
        title=document['title'],
        quantities=r
    ))
````

````python
with open('output.json', 'w') as f:
    json.dump(results, f, indent=2, ensure_ascii=False)
````
